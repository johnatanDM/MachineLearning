```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset =  keras.datasets.fashion_mnist
((imagens_treino, identificacoes_treino),(imagens_teste, identificacoes_teste)) = dataset.load_data()
print(imagens_treino.shape)
print(imagens_teste.shape)
```

```python
plt.imshow(imagens_treino[0])
plt.title(identificacoes_treino[0])
```

```python
print(identificacoes_treino)
print(identificacoes_treino.min())
print(identificacoes_treino.max())
```

```python
classes = ['Camiseta', 'Calça', 'Pullover', 'Vestido', 'Casaco', 
           'Sandalia', 'Camisa', 'Tenis', 'Bolsa', 'Bota']
```

```python
for imagem in range(10):
  plt.subplot(2,5, imagem + 1)
  plt.imshow(imagens_treino[imagem])
  plt.title(classes[identificacoes_treino[imagem]])

```

#Criando o nosso modelo

```python
modelo = keras.Sequential([
  keras.layers.Flatten(input_shape = (28, 28)),
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dense(10, activation=tensorflow.nn.softmax)
])

modelo.compile(optimizer='adam', 
               loss='sparse_categorical_crossentropy')
modelo.fit(imagens_treino, identificacoes_treino, batch_size=1)
```

```python
testes = modelo.predict(imagens_teste)
print('Resultado teste: ',np.argmax(testes[3]))
print('Numero da imagem de teste:', identificacoes_teste[3])
```

```python
perda_teste, acuracia_teste = modelo.evaluate(imagens_teste, identificacoes_teste, batch_size=1)
print("perda do teste: ", perda_teste)
print("acurácia teste: ", acuracia_teste)
```

```python
historico = modelo.fit(imagens_treino, identificacoes_treino, epochs=5, validation_split=0.2)
```

```python
historico.history
```

```python
def plot_acuracia(historico):
  plt.plot(historico.history['accuracy'])
  plt.plot(historico.history['val_accuracy'])
  plt.title('Acurácia por épocas')
  plt.xlabel('Épocas')
  plt.ylabel('Acurácia')
  plt.legend(['Treino', 'Avaliação'])
```

```python
plot_acuracia(historico)
```

```python
def plot_perda(historico):
  plt.plot(historico.history['loss'])
  plt.plot(historico.history['val_loss'])
  plt.title('Perda por épocas')
  plt.xlabel('Épocas')
  plt.ylabel('Perda')
  plt.legend(['Treino', 'Avaliação'])
```

```python
plot_perda(historico)
```

```python
modelo = keras.Sequential([
  keras.layers.Flatten(input_shape = (28, 28)),
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dropout(0.2),
  keras.layers.Dense(10, activation=tensorflow.nn.softmax)
])

modelo.compile(optimizer='adam', 
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

```python
historico = modelo.fit(imagens_treino, identificacoes_treino, epochs=5, validation_split=0.2)
```
